```python
import networkx as nx
```

```python
G = nx.DiGraph()
G_tree = nx.DiGraph()

for line in open('../level_k_lgt_generator/random_lgt_networks/network_4900.gr').readlines()[1:]:
    u = line.split(' ')[0]
    v = line.split(' ')[1]
    kind = line.rstrip('\n').split(' ')[2]
    G.add_edge(u,v,kind=kind)
    if kind=='tree':
        G_tree.add_edge(u,v)
```

```python
pos = nx.kamada_kawai_layout(G_tree)
```

```python
nx.draw_networkx_nodes(G,pos,node_size=3)
nx.draw_networkx_labels(G,pos,font_size=8)
nx.draw_networkx_edges(G, pos, edgelist=[(u,v) for (u,v) in G.edges if G[u][v]['kind']=='tree'],edge_color="tab:red")
nx.draw_networkx_edges(G, pos, edgelist=[(u,v) for (u,v) in G.edges if G[u][v]['kind']=='transfer'],edge_color="tab:blue")
```
