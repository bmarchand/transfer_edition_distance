```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import RegularPolygon
from matplotlib.collections import PatchCollection

# Example upper-triangular matrix
n = 6
A = np.triu(np.random.rand(n, n))

fig, ax = plt.subplots(figsize=(6, 6))
patches = []
colors = []

for i in range(n):
    for j in range(i, n):
        # Rotation transform: main diagonal lies horizontally
        x = i + j
        y = j - i
        # Add a diamond centered at (x, y)
        diamond = RegularPolygon((x, y), numVertices=4, radius=0.5, orientation=np.pi/4)
        patches.append(diamond)
        colors.append(A[i, j])

# Create a PatchCollection with colors mapped to matrix values
collection = PatchCollection(patches, array=np.array(colors), cmap='viridis', edgecolor='k')
ax.add_collection(collection)
ax.autoscale_view()
ax.set_aspect('equal')

# Add colorbar and labels
plt.colorbar(collection, ax=ax, label='Value')
ax.set_title("Upper Triangular Matrix (Tilted Heatmap)")
plt.show()
```
