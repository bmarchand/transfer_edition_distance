```python
fname = "results_nleaves50_nruns20_nreconcruns100"
```

```python
import matplotlib.pylab as plt
```

```python
nruns = 50
```

```python
data = []
data_vec = []
agg_values = []
cnt = 0
aggreg_mode = False
for line in open(fname).readlines():
    if line.find("num") >= 0 and line.find("aggregated") < 0:
        if cnt > 0:
            data.append(data_vec)
            data_vec = []
        cnt += 1
    if not aggreg_mode:
        if line.find("distance") >= 0:
            data_vec.append(float(line.split(" ")[3]))
    print(line)
    if line.find("aggregated") >=0:
        if not aggreg_mode:
            data.append(data_vec)
        aggreg_mode = True
        
    if aggreg_mode:
        if line.find("distance") >=0:
            agg_values.append(float(line.split(" ")[3]))
```

```python
fig, axs = plt.subplots(nrows=1,ncols=len(data),figsize=(20,2))

for i in range(len(data)):
    axs[i].hist(data[i])
    axs[i].axvline(x=agg_values[i],ymin=0,ymax=50,color='red')
    
```
